```python
import os
from pydub import AudioSegment

snippet_length = 5000
```

```python
genres = os.listdir('Database/Sorted Data/')
for genre in genres:
    #this keeps track of what snippet we're on so we can have unique names for each one
    s_counter = 1;
    for filename in os.listdir('Database/Sorted Data/' + genre + '/'):
        path = 'Database/Sorted Data/' + genre + '/' + filename
        song = AudioSegment.from_mp3(path)
        i = 0
        while i < len(song) - 500:
            snippet = song[i:i+snippet_length]
            s_name = genre+"_Snippet_"+str(s_counter)+'.mp3'
            snippet.export('Database/Sorted Data/' + genre + '/' + s_name, format="mp3")
            print("Created " + genre + " snippet " + str(s_counter))
            s_counter = s_counter+1
            if os.path.isdir('Database/Sorted Data Snippets/' + genre):
                os.rename('Database/Sorted Data/' + genre + '/' + s_name, 'Database/Sorted Data Snippets/' + genre + '/' + s_name)
            else:
                os.makedirs('Database/Sorted Data Snippets/' + genre)
                os.rename('Database/Sorted Data/' + genre + '/' + s_name, 'Database/Sorted Data Snippets/' + genre + '/' + s_name)
            i = i + snippet_length
```
