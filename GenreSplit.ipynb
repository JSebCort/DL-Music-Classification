```python
import os
import csv
import IPython.display as ipd
import numpy as np
import pandas as pd
import glob

def cleanFilename(filename):
    parts = filename.split(".")
    filename = parts[0]
    while list(filename)[0] == '0':
        filename = filename[1:]
    return filename
```

```python
metadata = list(csv.reader(open('Database/fma_metadata/tracks.csv')))
#genre stored in 40th column of csv file
#track id stored in 0th column
#metadata = pd.read_csv('Database/fma_metadata/tracks.csv', low_memory=False, usecols = ['Unnamed: 0', 'track.7'])
folder_names = []
for i in range(0,156):
    temp = str(i)
    while len(temp) < 3:
        temp = '0'+temp
    folder_names.append(temp)

for folder in folder_names:
    for filename in os.listdir("Database/fma_small/" + folder + "/"):
        track_id = cleanFilename(filename)
        genre = ""
        for i in range (0,len(metadata)):
            if metadata[i][0] == track_id:
                genre = metadata[i][40]
                i = len(metadata) + 1
        if os.path.isdir('Database/Sorted Data/' + genre):
            os.rename('Database/fma_small/' + folder + "/" + filename, 'Database/Sorted Data/' + genre + "/" + filename)
        else:
            os.makedirs('Database/Sorted Data/' + genre)
            print("Made directory for " + genre)
            os.rename('Database/fma_small/' + folder + "/" + filename, 'Database/Sorted Data/' + genre + "/" + filename)
        print("Currently moving " + filename + " to directory: " + genre)
```
